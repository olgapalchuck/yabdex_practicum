<h3> XYZ–анализ </h3><br>
 - для определения стабильности продаж товара за определенный период. Полезен для управления ассортиментом и поставками товаров, организации работы с поставщиками, позволяет разделить товары по категориям и выделить для них место на складе, уровень запасов и организацию доставки. 
 - Критерии для анализа - количество проданного товара за определенный период, спрос на товар, количество покупок.<br>
<h4> Методика анализа </h4>
 - для анализа берем все товарные позиции, которые есть в данных<br>
 - период анализа - год<br>
 - рассчитываем среднеарифметическое значение количества продаж каждого товара за год по месяцам<br>
 - считаем среднеквадратическое отклонение<br>
 - определяем коэффициент вариации для каждого товара, это будет относительный показатель<br>
 - сортируем полученные значения коэффициента вариации от меньшего к большему<br>
 - расшифровка значений: <br>
   - Категория Х, в которую попадают товары с минимальным колебанием продаж, характеризующиеся стабильной величиной потребления и высокой степенью прогнозирования до 0,1-0,2.<br>
   - Категория Y, в которую попадают товары со средним колебанием продаж от 0,2 до 0,6, с сезонными колебаниями и средними возможностями их прогнозирования.<br>
   - Категория Z, в которую попадают товары с резкими колебаниями продаж от 0,6 и выше, с нерегулярным потреблением и непредсказуемыми колебаниями, поэтому, спрогнозировать их спрос невозможно.<br>
 - визуализируем данные

<h3> АВС–анализ </h3><br>

В основе рассматриваемого анализа лежит принцип Парето, который акцентирует внимание на том, что 20% усилий обеспечивают 80% результата. Остальные 20% результата достигаются через 80% усилий. Критерии анализа: фактический объем продаж, фактическая прибыль.<br>
<h4> Методика анализа </h4>
 - для анализа берем все товарные позиции, которые есть в данных<br>
 - период анализа - год<br>
 - рассчитываем прибыль от реализации каждого товара за год<br>
 - сортируем товар по убыванию продаж<br>
 - расчитываем долю каждого товара и совокупный процент<br>
 - расшифровка значений: <br>
 - А – лидеры, за счет которых формируется основная прибыль. Эти позиции не рекомендуется изменять или исключать.Можно инвестировать в рекламу, направленную на раскрутку этих позиций.<br>
 - В – пул, который обеспечивает средние продажи. Требует меньше инвестиций на продвижение, но может отличаться нестабильным спросом.<br>
 - С – товары, которые занимают последние позиции. Их необходимо исключить, улучшить или оставить, если они играют важную роль в позиционировании.<br>

<h4>2. Проверка количественных данных на наличие выбросов.

```python
df = df.drop(columns='last_id', axis=1).reset_index()
df['sum_sale'] = df['quantity'] * df['price']

df['quantity'].max()
df[df['quantity'] == 1000]

print(np.percentile(df['price'], [80, 90, 95]))
print(np.percentile(df['sum_sale'], [80, 90, 95]))


labels = ['1/6', '1/3', '.5', '2/3', '5/6', '1']
df['quantile'] = pd.qcut(df['price'], q=[0, 1/6, 1/3, .5, 2/3, 5/6, 1], labels=labels)

for label in labels:
    
    df_quant = df.query("quantile == @label")
    fig, (ax1, ax2) = plt.subplots(1, 2)
    ax1.hist(df_quant['price'])
    ax2.violinplot(df_quant['price'])
    ax1.set_title(f"percentile {label}")
    plt.show()
```

Как мы видим из графиков цены за единицу товара, у нас много товаров со стоимостью  около 40, 120,  150 и 200 уе (плюс-минус 10 уе). 5/6 всего товара дешевле 750 уе., остальной товар имеет разброс от 750 до 15 000 уе - т е на каждый дальнейший процент шага у нас будет нарастать разброс в цене, поэтому данные будут прижаты к левому краю гистограммы и иметь длинный хвост. 

```python
labels = ['1/6', '1/3', '.5', '2/3', '5/6', '1']
df_cheque = df.groupby('order_id', as_index=False).agg({'sum_sale': 'sum'})
df_cheque['quantile'] = pd.qcut(df_cheque['sum_sale'], q = [0, 1/6, 1/3, .5, 2/3, 5/6, 1], labels=labels)

for label in labels:
    
    df_quant_ch = df_cheque.query("quantile == @label")
    fig, (ax1, ax2) = plt.subplots(1, 2)
    ax1.hist(df_quant_ch['sum_sale'])
    ax2.violinplot(df_quant_ch['sum_sale'])
    ax1.set_title(f"percentile {label}")
    plt.show()
```

Мы видим ситуацию со средним чеком аналогичную стоимости за единицу товара: 5/6 всех средних чеков будут меньше 2 300 уе, оставшаяся 1/6 будет иметь разброс от 2 300 до 650 000 уе. - это еще больший разброс на каждый процент шага.<br>
Строить гистограмму по количеству не будем - посмотрим на шагах перцентилей в 1%, как меняется количество товара одного наименования в чеке - подавляющее большинство чеков - 81% - по одному товару, 18% - от 2 до 25 товаров и только 1% чеков больше 100 товаров.

<h5>Блок автоматической группировки данных по задонному столбцу и построения графиков</h5>

```python
def per_day(data):

    ''' Фукция подсчета среднего количества покупок, покупателей, 
    приобретенных наименований продуктов и количества товара, и выручки''' 
    
    per_day = data.groupby(['month_day']).agg({'order_id': 'nunique',\
                                      'customer_id': 'nunique',\
                                      'product': 'nunique',\
                                      'quantity': 'sum',\
                                     'sum_sale': 'sum'})
    
    print(f"За период времени, равный {len(per_day)} дн")
    print(f"Среднее количество в день покупок {per_day['order_id'].mean()}")
    print(f"Среднее количесвтво в день уникальных посетителей {per_day['customer_id'].mean()}")
    print(f"Среднее количество в день приобретенных наименований продуктов {per_day['product'].mean()}")
    print(f"Среднее количество в день приобретенного количества товара {per_day['quantity'].mean()}")
    print(f"Среднее количество в день выручки {per_day['sum_sale'].mean()}")
```

```python
def cheque_per_column (data, column):
    
    """функция для группировки датасета по заданному столбцу 
    возвращает средний и медианный чек, сгруппированные по заданному столбцу 
    median_cheque - медианный чек, mean_cheque - средний чек"""
    

    data_sum_cheque = data.groupby([column, 'order_id'])\
        .agg({'sum_sale': 'sum'})\
        .rename(columns=({'sum_sale': 'sum_one_cheque'}))\
        .reset_index()

    data_median_cheque = data_sum_cheque.groupby(column)\
        .agg({'sum_one_cheque': 'median'})\
        .rename(columns=({'sum_one_cheque': 'median_cheque'}))\
        .reset_index()
    
    data_mean_cheque = data_sum_cheque.groupby(column)\
        .agg({'sum_one_cheque': 'mean'})\
        .rename(columns=({'sum_one_cheque': 'mean_cheque'}))\
        .reset_index()
    
    data_mean_cheque['mean_cheque'] = (data_mean_cheque['mean_cheque']).round(2)
    
    data_ = data_median_cheque.merge(data_mean_cheque, on=column)

    return data_
```

```python
def gr_sale_per_month(data, column):
    
    """функция для группировки датасета по задонному столбцу 
    возвращает общую сумму продаж, количество покупок, уникальных клиентов, средний и медианный чеки, сгруппированные по задоному столбцу
    total_revenue - общая выручка за мес, тыс у е, total_buy - общее количество покупок, last_month - выручка в тыс за прошлый мес, 
    grow_per_month_pr - рост выручки данного месяца относительно предыдущего в %, 
    mean_cheque - средний чек, total_buyer - общее количество уникальных покупателей,
    median_cheque - медианный чек, mean_cheque - средний чек"""

    if column == 'month_year':
        data_ = data.groupby(column)\
            .agg({'sum_sale': 'sum',\
                'order_id': 'nunique',\
                'customer_id': 'nunique'})\
            .rename(columns=({'sum_sale': 'total_revenue',\
                            'order_id': 'total_buy',\
                            'customer_id': 'total_buyer'}))\
            .reset_index()
    
        data_['total_revenue'] = data_['total_revenue'] / 1000 
        data_['last_month'] = np.roll(data_['total_revenue'], 1)
        data_['grow_per_month_pr'] = (data_['total_revenue'] / data_['last_month']) - 1
        data_['grow_per_month_pr'] = (data_['grow_per_month_pr'] * 100).round(2)
        
        data1_ = cheque_per_column(data, column)
        data_ = data_.merge(data1_, on=column)
        
        return data_
    
    else:
        data_ = data.groupby(column)\
            .agg({'customer_id': 'nunique',\
                'order_id': 'nunique'})\
            .rename(columns=({'order_id': 'total_buy',\
                            'customer_id': 'total_buyer'}))\
            .reset_index()
        
        data1_ = cheque_per_column(data, column)
        data_ = data_.merge(data1_, on=column)
        
        return data_
    
```

<h4>Количество товаров в одном чеке</h4>
Посмотрим, как меняется среднее (по медиане везде единица) наименований товаров в одном чеке по месяцам, дням недели и часам.<br>
Покупку вантузов мы снова уберем.

```python
def product_in_cheque(data, column):
    
    ''' Функция группирует датафрейм по заданному столбцу, счиатем средний чек
    и возращает данные в виде графиков'''
    
    data_gr = data.groupby([column, 'order_id']).agg({'product': 'count'}).reset_index()
    data_gr = data_gr.groupby(column).agg({'product': 'mean'}).reset_index()
    data_gr['product'] = (data_gr['product']).round(2)
    
    plt.figure(figsize=(7, 3))
    plt.plot(data_gr[column], data_gr['product'], label='n_product_in_cheque')
    plt.xticks(rotation='vertical')
    plt.grid()
    plt.xlabel(column)
    plt.ylabel('количество')
    plt.title(f' График изменения количества уникальных товаров \n в одном чеке по {column}', fontsize=12)
    plt.show()
```

```python
columns = ['month_year', 'day_week', 'hour']
for column in columns:
    product_in_cheque(df_aut, column)
```

В среднем, среднее количество товаров в чеке колеблется около двух. Это возможная точка роста - необходимо предлагать перед оплатой товара дополнительные товары из данной или смежной категорий, или давать скидку на третий и более товар в чеке. Июнь 2019 г был уникальный месяц - с высокой общей выручкой, высоким средним чеком и самым большим средним количеством наименований в товаре - 8 шт. По субботам заказов мало, но берут больше товара в одном чеке. Также в пн и пт берут больше товаров в одном чеке - около 2.3 ед. Если рассматривать по часам, то больше всего товаров берут в 9, 12, 16 часов.

Наиболее активны пользователи с 8 утра до 21 вечера, пик активности - 9 утра - 15 дня. Средний чек максимален в 11 утра, и с 16 до 18, настройку рекламы стоит делать на период с 10.30 до 14.30, и с 15.30 до 17.30, чтобы превлечь максимальное количество пользователей.   <br>

## XYZ-анализ <a class="anchor" id="four-bullet"></a>
Теперь классифицируем товар по частоте продаж.<br><br>
<h5>Описание кода</h5>
Для каждого наименования товара df['product'].unique() считаем общее количество проданных за год единиц и делим на 13 - количество месяцев в выборке - находим среднемесячную реализацию. Отдельно сохраняем ежемесяную реализацию товара в sale_per_months.<br>
Затем считаем среднеквадратическое отклонение - показатель рассеивания в продажах по выбранному значению. Чем больше среднеквадратическое отклонение - тем больше разница между среднегодовыми и среднемесячными продажами. Среднеквадратическое отклонение считаем по формуле:

Далее для каждого товара считаем коэффициент вариации - относительный показатель. Чем выше коэффициент вариации - тем нестабильнее продажи.

Категория Х (мин колебания продаж, стабильный спрос и высокая степень прогнозирования) - до 0,1-0,2.<br>
Категория Y (товары со средним колебанием продаж, с сезонными колебаниями и средними возможностями их прогнозирования) от 0,2 до 0,6.<br>
Категория Z (товары с резкими колебаниями продаж, с нерегулярным потреблением и непредсказуемыми колебаниями) - от 0,6 и выше. 

```python
names = df['product'].unique()
dict_koef = []
for name in names:
    df_name = df.query('product == @name')
    mean_sale_year = (df_name['quantity'].sum() / 13).round(2)
    sale_per_months = []
    for date in df['month_year'].unique():
        data_ = df_name.query('month_year == @date')
        sale_per_month = data_['quantity'].sum()
        sale_per_months.append(sale_per_month)
        standart_dev1 = []
        for i in sale_per_months:
            standart_dev1.append(((i - mean_sale_year)**2))
            standart_dev = math.sqrt((sum(standart_dev1)) / 13)
            koef_var = (standart_dev / mean_sale_year).round(2)
    dict_koef.append({'product': name,  
                     'koef_var': koef_var})
```

```python
pd_koef = pd.DataFrame(dict_koef, columns=['product', 'koef_var'])
pd_koef = pd_koef.sort_values(by='koef_var')
pd_koef.head()
```

```python
def xyz_group (row):
    if row['koef_var'] <= 0.2:
         return "X"
    if  0.2 < row['koef_var'] < 0.6:
        return "Y"
    else:
        return "Z"
```

```python
pd_koef['xyz_category'] = pd_koef.apply(xyz_group, axis=1)
pd_koef.head()
```

Весь товар попадает в категорию Z - товары с резкими колебаниями продаж, с нерегулярным потреблением и непредсказуемыми колебаниями. Возможно, данных для XYZ-анализа было недостаточно, так как хозяйственные товары не пользуются таким широким спросом, как например, продукты питания, и необходимо брать выборку данных до 2-3 лет, чтобы получить доставерный результат. В данном случае можно попытаться провести XYZ-анализ по категориям товаров: сумки-тележки, одеяла, стремянки и т д.<br><br>

```python
X = np.array(df_not_wt_gr['sum_sale'])
Y = np.array(df_wt_gr['sum_sale'])

alpha = 0.05
results = st.ttest_ind(Y, X, equal_var=False)

print('p-value:', results.pvalue)

if results.pvalue < alpha:
    print('отвергаем нулевую гипотезу')
else:
    print('не получилось отвергнуть нулевую гипотезу')
```

Хотя средний чек в двух выборках отличается, вероятность получить такой результат случайно почти 16%, поэтому средний чек во вт и пт не отличается от среднего чека в другие дни.
